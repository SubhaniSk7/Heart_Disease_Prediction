```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('heart.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data["target"].value_counts()
```

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(data, hue='target', vars=["age", "sex", "cp", "trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal"],size=5, palette="husl");
plt.show()
```
